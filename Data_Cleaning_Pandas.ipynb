```python
import pandas as pd
```

```python
# Daten aus Excel-Datei importiert
df = pd.read_excel('path\Customer Call List.xlsx')
df
```

```python
# Duplikate entfernen und wird als df gespeichert
df = df.drop_duplicates()
df
```

```python
# Nicht benötigte Spalten löschen und wird als df gespeichert
df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
# Spalte Last_Name wird bereinigt. strip() entfernt nicht benötigte Zeichen links und rechts vom Nachnamen
# Wichtig ist, dass es nicht als df, sondern als df['Last_Name'] gespeichert wird, denn nur die Spalte wird hier bearbeitet
df['Last_Name'] = df['Last_Name'].str.strip('123._/')
df
```

```python
# Spalte Phone_Number standardisieren, erst alles in Zahlen umwandeln, dann wieder mit replace() - einsetzen
# [^a-zA-Z0-9] bedeutet, dass alles ersetzt wird, außer die Buchstaben klein a-z und groß geschrieben A-Z und Zahlen 0-9
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '')
df
```

```python
# Erst alles in string umwandeln, da mit float Lambda nicht funktioniert
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
# Mit Lambda wird - eingefügt
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
# Alle NAs werden ersetzt durch Leerzeichen
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
df
```

```python
# Spalte Address bleibt erhalten, aber die Daten werden auf drei neue Spalten aufgeteilt und hinzugefügt
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', 2, expand = True)
df
```

```python
# Spalte Paying Customer wird standardisiert
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')

df['Paying Customer'] = df['Paying Customer'].str.replace('Y', 'Yes', regex = False)
df['Paying Customer'] = df['Paying Customer'].str.replace('N', 'No', regex = False)
df
```

```python
# Spalte Do_Not_Contact wird standardisiert
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Y', 'Yes', regex = False)
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('N', 'No', regex = False)
df
```

```python
# Ganze Tabelle mit String_NAs werden mit Leerzeichen ersetzt
df = df.replace('N/a', '')
df = df.replace('NaN', '')
# Tatsächliche Nulls werden mit Leerzeichen ersetzt
df = df.fillna(value='')
df
```

```python
# Filterung: Wenn Do_Not_Contact Yes ist, dann sollen die Zeilen gelöscht werden
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Yes':
        df.drop(x, inplace = True)

df
```

```python
# Filterung: Wenn Phone_Number leer ist, dann sollen die Zeilen gelöscht werden
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace = True)

df
# Am Ende werden nur noch Zeilen angezeigt, wo Phone_Number vorhanden und Do_Not_Contact mit No hinterlegt sind
```

```python
# Index neu anlegen, zeigt die Anzahl der Kontakte, die man anrufen möchte
df = df.reset_index(drop = True)
df
```
