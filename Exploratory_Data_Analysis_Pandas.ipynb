```python
# Ich will nicht nur Daten analysieren, sondern auch visualisieren, dafür werden folgende Pakete importiert
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Tabelle aus der CSV-Datei importiert
df = pd.read_csv('path\world_population.csv')
df
```

```python
# 2 Nachkommastellen einstellen
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# Überblick über die Spalten, Nulls, Datentypen
df.info()
```

```python
# Überblick über die Statistik, z.B. ist die Anzahl der Bevölkerung über die Jahre gestiegen
df.describe()
```

```python
# Wie viele Nulls haben welche Spalten? Für Datenbereinigung, -analyse, -visualisierung von Bedeutung.
df.isnull().sum()
```

```python
# Anzahl der einzigartigen Werte, z.B. es sind nur 6 Kontinente eingetragen
df.nunique()
```

```python
# Top10 der Länder mit den höchsten Bevölkerungszahl abwärts sortiert
df.sort_values(by = '2022 Population', ascending = False).head(10)
```

```python
# Korrelation, z.B. Korrelation Population und Area sind eher niedrig, was bedeutet, dass es große Länder mit hohem Bevölkerungszahl gibt, aber auch kleine Länder mit hohem Bevölkerungszahl oder andersrum
df.corr()
```

```python
# Erstelle Heatmap zu der Korrelation und Anpassung vo Graph und Schriftgröße
heatmap = sns.heatmap(df.corr(), annot = True, annot_kws={"size": 18})
plt.rcParams['figure.figsize'] = (20, 7)
heatmap.set_xticklabels(heatmap.get_xticklabels(), fontsize=18)
heatmap.set_yticklabels(heatmap.get_yticklabels(), fontsize=18)
plt.show()
```

```python
# Gruppierung nach Kontinent und Sortierung abwärts nach Durchschnittsbevölkerungsanzahl
df.groupby('Continent').mean().sort_values(by = '2022 Population', ascending = False)
```

```python
# Ich will mir Kontinent Oceania und die Länder ansehen. Oceania ist eine Ansammlung von Australien und mehrere Inseln
df[df['Continent'].str.contains('Oceania')]
```

```python
# Ich will nur die Bevölkerungszahl über die Jahre visualisieren und Reihenfolge geändert
df2 = df.groupby('Continent')[df.columns[5:13][::-1]].mean().sort_values(by = '2022 Population', ascending = False)
# X und Y werden transponiert
df3 = df2.transpose()
df3
```

```python
# Liniendiagramm: In den 90ern ist die Bevölkerungszahl in Asien explodiert
df3.plot()
```

```python
# Boxplot: Sehr hohe Anzahl an Ausreißer. Untersuchen, warum die Länder solch hohe Ausreißer abbilden, sind es valide Zahlen?
df.boxplot()
```
